import pandas as pd
df = pd.read_csv('../dataset/dataset.csv')
df.head()

```python
df.describe()
```

```python
stats = df.groupby('class').count().rename(columns={'text': 'count'})
plot = stats.plot(kind='bar')
```

```python
russia_stats = df[df['text'].str.lower().str.contains('([^a-z]|^)(russian?)([^a-z]|$)')].groupby('class').count().rename(columns={'text': 'count'})
russia_stats
```

```python
percent = stats.join(russia_stats, rsuffix='_russia')
percent.insert(loc=len(percent.columns), column='percent', value=(percent['count_russia'] / percent['count'] * 100))
percent = percent.sort_values(by='percent', ascending=False)
percent
```

```python
import re
def splitDataFrameList():
    def splitListToRows(row):
        split_row = row['text'].lower().split()
        for s in split_row:
            word = re.sub(r"(^[^a-z]+)|([^a-z]+$)", "", s)
            if (word != ""):
                new_rows.append({'word': word, 'class': row['class']})
    new_rows = []
    df.apply(splitListToRows,axis=1)
    new_df = pd.DataFrame(new_rows)
    return new_df

words = splitDataFrameList()
```

```python
word_counts = words.groupby(['class', 'word'])['word'].count().reset_index(name="count").sort_values(by='count', ascending=False)

word_counts
```

```python

top_word_counts = word_counts.groupby('class').head(10).reset_index(drop=True).sort_values(by=['class', 'count'], ascending=[True, False])

top_word_counts

```

```python
stop_words = set()
with open('../terms/bbc.terms') as f:
    stop_words = set(f.read().split("\n"))

filtered_rows = []
def filterRow(row):
    d = row.to_dict()
    word = d['word']
    if (not (word in stop_words)):
        filtered_rows.append(d)
        
word_counts.apply(filterRow, axis=1)


filtered_words = pd.DataFrame(filtered_rows).groupby('class').head(10).reset_index(drop=True).sort_values(by=['class', 'count'], ascending=[True, False])

filtered_words
```
